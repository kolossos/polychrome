```python
im2=np.zeros((8,18),np.uint8)
im_list = []

for iii in range(174):
    first=True
    print(iii)
    for ii in range(zz.shape[0]):
        print(zz[ii,iii],end=";")
        i=(15.*digits.images[zz[ii,iii]]).astype("uint8")
        if first:
            ims=i
            first=False
        else:
            ims=np.concatenate((ims,im2),axis=1)
            ims=np.concatenate((ims,i),axis=1)
    im_list.append(Image.fromarray(ims))
    im = Image.fromarray(ims)
    im.save("MNIST3/"+str(iii)+".png")

img = im_list[0]
imgs = im_list[1:]
img.save("MNIST3/output_webp2.webp", save_all=True, append_images=imgs, duration=90, loop=0, optimize=False, disposal=2, lossless=True)
```
